```python
# Importing libraries
import pandas as pd
import numpy as np
import requests
from io import StringIO
```

```python
""" Loading the data """

# Read the data
url = "https://pynative.com/wp-content/uploads/2019/01/Automobile_data.csv"

# Send request with headers
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)

# Load into pandas
df = pd.read_csv(StringIO(response.text))
df.info()
```

```python
# Print the first and last five columns
df.head()
```

```python
# Identify missing values and choose a method to replace them
print(df.isnull().sum())
# Select only numeric columns to calculate the mean
numeric_cols = df.select_dtypes(include=np.number)
df['price'] = df['price'].fillna(numeric_cols.mean()['price'])
# Print the missing values to see if there is any remaining with numerical columns
print("\nPrinting again after handling missing values\n")
print(df.isnull().sum())
```

```python
# Print the most expensive car, the company name, and the price.
df[df['price'] == df['price'].max()][['company', 'price']]
```

```python
# Count total cars per company
df['company'].value_counts()
```

```python
# Find each companyâ€™s highest-priced car
df.groupby('company')['price'].max()
```

```python
# Find the average mileage of each car-making company
df.groupby('company')['average-mileage'].mean()
```

```python
# Sort all cars by the price column
df.sort_values(by='price', ascending=False)
```
